## Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\akars\OneDrive\Desktop\6th Sem\Machine Learning\Lab\DataSets\Diwali.csv", encoding="latin1")
print(df.head())
```

```python
df.shape 
```

```python
healthcare= df[df['Occupation'].str.contains('Healthcare')]
healthcare_count= healthcare.shape[0]
print(f"Total number of healthcare occupations:{healthcare_count}")
```

## Graph Visualization

```python
categories = ['Healthcare Occupations', 'Other Occupations']
counts = [healthcare_count, len(df) - healthcare_count]

plt.bar(categories, counts, color=['purple', 'yellow'])
plt.title('Distribution of Healthcare Occupations')
plt.xlabel('Occupation Type')
plt.ylabel('Count')
plt.show()
```
